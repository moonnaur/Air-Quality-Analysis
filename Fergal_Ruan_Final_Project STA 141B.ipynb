```python
# API token
token = 'fcd58a32ea3042b6bd1e4af607da16f07fd9e9fb'

# Define the new latitude/longitude range
regions = {
    'USA_Wide': '25,-140,50,-70'  
}

# Base URL for the WAQI API
base_url = "https://api.waqi.info/map/bounds/"

# Create an empty list to store the data
data_list = []

# Iterate over the defined region to fetch air quality data
for region, latlngbox in regions.items():
    # Construct the request URL
    url = f"{base_url}?latlng={latlngbox}&token={token}"
    
    # Send a GET request to the API
    response = requests.get(url)
    
    # Process the response
    if response.status_code == 200:
        data = response.json()
        if data['status'] == 'ok':
            # Extract data for each station and add it to the list
            for station in data['data']:
                station_data = {
                    'Station': station['station']['name'],
                    'Latitude': station['lat'],
                    'Longitude': station['lon'],
                    'AQI': station['aqi']
                }
                data_list.append(station_data)
        else:
            print(f"Error retrieving data for {region}: {data['data']}")
    else:
        print(f"Failed to retrieve data for {region}: Status code {response.status_code}")

# Convert the list of dictionaries to a DataFrame
air_quality_data = pd.DataFrame(data_list)
```

```python
def fetch_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract city and state from URL
    parts = url.split('/')
    state = parts[-2] if len(parts) > 2 else None
    city = parts[-1] if len(parts) > 1 else None

    # Population extraction
    population_label = soup.find('div', class_='number-big-text', string='Population')
    population_number = population_label.find_next_sibling('div')['data-count'] if population_label else None
    
    # Energy consumption extraction
    energy_label_div = soup.find('div', class_='number-big-text', string='Energy consumption')
    energy_consumption = None
    if energy_label_div:
        energy_value_div = energy_label_div.find_next_sibling('div')
        if energy_value_div:
            energy_span = energy_value_div.find('span', class_='number-big')
            energy_consumption = energy_span['data-count'] if energy_span else None
    
    # GHG emissions extraction
    ghg_label_div = soup.find('div', class_='number-big-text', string='GHG emissions')
    ghg_emissions = None
    if ghg_label_div:
        ghg_value_div = ghg_label_div.find_next_sibling('div')
        if ghg_value_div:
            ghg_span = ghg_value_div.find('span', class_='number-big')
            ghg_emissions = ghg_span['data-count'] if ghg_span else None

    return {
        'City': city.replace('-', ' ').title(),  # Convert URL format to normal city name
        'State': state.replace('-', ' ').title(),  # Convert URL format to normal state name
        'Population': population_number, 
        'Energy Consumption': energy_consumption,
        'GHG Emissions': ghg_emissions
    }

data = []
with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
    futures = [executor.submit(fetch_data, url) for url in url_list]  # Change here for full list
    for future in concurrent.futures.as_completed(futures):
        data.append(future.result())

# Convert list to DataFrame
city_data = pd.DataFrame(data)
```

```python
# Remove rows where Population, Energy Consumption, or GHG Emissions data are missing.
city_data.dropna(subset=['Population', 'Energy Consumption', 'GHG Emissions'], inplace=True)

# Reset the DataFrame index to account for the dropped rows and ensure continuity.
city_data = city_data.reset_index(drop=True)

# Convert the 'Population' column to numeric type, coercing any errors (non-numeric values) into NaN.
city_data['Population'] = pd.to_numeric(city_data['Population'], errors='coerce')

# Convert the 'Energy Consumption' column to numeric type and round to 2 decimal places, 
# coercing errors into NaN. This ensures data uniformity and precision.
city_data['Energy Consumption'] = pd.to_numeric(city_data['Energy Consumption'], errors='coerce').round(2)

# Convert the 'GHG Emissions' column to numeric type and round to 2 decimal places,
# also coercing errors into NaN for consistency and precision.
city_data['GHG Emissions'] = pd.to_numeric(city_data['GHG Emissions'], errors='coerce').round(2)

city_data
```

```python
# Aggregate city_data by 'State' and sum numerical columns. This combines all cities within each state 
# and totals their population, energy consumption, and GHG emissions for statewide analysis.
state_data = city_data.groupby('State').sum().reset_index()

# Drop the 'City' column from state_data as it's no longer needed after aggregation. The focus is on state-level data.
state_data = state_data.drop(columns=['City'])
```
